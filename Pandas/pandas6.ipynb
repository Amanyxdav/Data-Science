```python
import pandas as pd
import numpy as np
df = pd.read_csv('weather_data_pandas6.csv')
df
```

## df.replace([-99999, -88888],np.NaN) is used for replacing NaN values in Pandas

```python
new_df = df.replace([-99999,-88888, 0],np.NaN)
new_df
```

```python
new_df = df.replace ({
    'temperature':[-99999,-88888],
    'windspeed': [-99999,-88888],
    'event':'0'
},np.NaN)
new_df
```

```python
new_df = df.replace({
    -99999: np.NaN,
    -88888: np.NaN,
    '0':'Sunny'
    
})
new_df
```

```python
new_df = new_df.replace('[A-Za-z]',' ',regex = True)
new_df
```

```python
new_df = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]',
    'event':'0'
},'',regex = True)
new_df
```

```python
new_df = df.replace({
    'event':'0'
    
},'  * No Event*  ')
new_df
```
